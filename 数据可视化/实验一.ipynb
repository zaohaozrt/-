```python
# ————面向对象的方式
import numpy as np
import matplotlib.pyplot as plt
x_data = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y_sin, y_cos = np.sin(x_data), np.cos(x_data)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x_data, y_sin)
ax.plot(x_data, y_cos)
plt.show()
```
