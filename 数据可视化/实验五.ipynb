```python
import numpy as np
import matplotlib.pyplot as plt
# 数据
x = np.arange(6)
man = [85.5, 91, 72, 59,66,55]
woman = [94,82,89.5,62,49,53]
bar_width = 0.3
plt.ylabel('平均成绩（分）')
# 绘图 x 表示 从那里开始
plt.title('高二男生、女生的平均成绩')
plt.bar(x, man, bar_width)
plt.bar(x+bar_width, woman, bar_width)
tick_label=["语文", "数学", "英语","物理", "化学", "生物"]
plt.xticks(x+bar_width/2,tick_label)
plt.axhline(88.5,linewidth=1.0,label='全体平均成绩')
kinds = ['全体平均成绩','男生','女生']
plt.legend(kinds, loc='upper left', bbox_to_anchor=[0.67, 1])
for i in range(6):
    plt.annotate(man[i],(x[i]-bar_width/2,man[i]))
    plt.annotate(woman[i],(x[i]+bar_width/2,woman[i]))
# 展示图片
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
# 绘制柱形图
subject=["语文","数学","英语","物理","化学",'生物']
man = [85.5,91,72, 59,66,55]
woman = [94,82,89.5,62,49,53]
ind=[x for x, _ in enumerate(subject)]
plt.bar(ind, man, width=0.5,  color='green') 
plt.bar(ind, woman, width=0.5,  color='orange', bottom=man) 
plt.xticks(ind,subject)
bar_width = 0.3
plt.ylabel('平均成绩（分）')
# 绘图 x 表示 从那里开始
plt.title('高二男生、女生的平均成绩')
plt.axhline(88.5,linewidth=1.0,label='全体平均成绩')
kinds = ['全体平均成绩','男生','女生']
plt.legend(kinds, loc='upper left', bbox_to_anchor=[0.67, 1])
for i in range(6):
    plt.annotate(man[i],(x[i]-bar_width/2,man[i]))
    plt.annotate(woman[i],(x[i]-bar_width/2,woman[i]+man[i]))
# 展示图片
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_data = np.linspace(0,2, 200, endpoint=True)
y1, y2,y3 = x_data, x_data**2,x_data**3
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x_data, y1)
ax.plot(x_data, y2)
ax.plot(x_data, y3)
plt.xlabel('x轴')
plt.ylabel('y轴')
kinds = ['y=x','y=x^2','y=x^3']
plt.legend(kinds, loc='upper left', bbox_to_anchor=[0, 1])
plt.grid(b=True, linewidth=0.4)
plt.title('不同曲线对比')
plt.show()
```
