```python

import matplotlib.pyplot as plt
import numpy as np
# 绘制柱形图
contries=["usa", "uk","cn","rus","ger"]
y_a = np.array([38,17, 26, 19, 15])
y_b = np.array([37,23, 18, 18, 10])
y_c = np.array([46,27, 26, 19, 17])
ind=[x for x, _ in enumerate(contries)]
plt.bar(ind, y_c, width=0.5, label='golds', color='green', bottom=y_a+y_b) 
plt.bar(ind, y_b, width=0.5, label='silvers', color='orange', bottom=y_a) 
plt.bar(ind, y_a, width=0.5, label='bronzes', color='royalblue') 
plt.xticks(ind,contries)
plt.xlabel("contries")
plt.show()
```
