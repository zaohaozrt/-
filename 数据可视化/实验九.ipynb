```python
#大数据202王申政 2029730224
import numpy as np
import matplotlib.pyplot as plt
def koch_snowflake(order, scale=10):
    def _koch_snowflake_complex(order):
        if order == 0:
        # 初始三角形
            angles = np.array([0, 120, 240]) + 90
            return scale / np.sqrt(3) * np.exp(np.deg2rad(angles) * 1j)
        else:
            ZR = 0.5 - 0.5j * np.sqrt(3) / 3
            p1 = _koch_snowflake_complex(order - 1) # 起点
            p2 = np.roll(p1, shift=-1) # 终点
            dp = p2 - p1 # 连接向量
            new_points = np.empty(len(p1) * 4, dtype=np.complex128)
            new_points[::4] = p1
            new_points[1::4] = p1 + dp / 3
            new_points[2::4] = p1 + dp * ZR
            new_points[3::4] = p1 + dp / 3 * 2
            return new_points
    points = _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    return x, y
x, y = koch_snowflake(order=2)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.fill(x, y, facecolor='lightsalmon', edgecolor='orangered', linewidth=3)
plt.show()
```

```python
#大数据202王申政 2029730224
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family']='simhei'
plt.rcParams['axes.unicode_minus']=False
x = np.linspace(0, 10, 11)
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1, 9.9, 13.9, 15.1, 12.5]

# 拟合曲线，并估计其y值及误差。
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b#曲线
y_err = x.std() * np.sqrt(1/len(x) +(x - x.mean())**2 / np.sum((x - x.mean())**2))#误差

# 绘制拟合线y_est，使用线性短虚线
plt.plot(x,y_est,linestyle=':')

# 填充置信带区域，y_est +- y_err，透明度0.2 。置信区域位于y_est - y_err, y_est + y_err两条线之间的区域
plt.fill_between(x,y_est - y_err,y_est + y_err,alpha=0.2)

# 绘制数据点：(x,y)，标记元素：圆形，颜色：绿色
plt.scatter(x,y, marker="o",color="g")

#添加标题，拟合曲线置信带案例，字体大小：20，颜色：红色，对齐方式：右对齐
plt.title("拟合曲线置信带案例",fontsize=20,loc='right',color="r")

plt.show()
```
